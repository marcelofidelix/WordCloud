```python
texto = '''
No mesmo dia, o ministro da Segurana Pblica, Raul Jungmann, disse que a PF vai investigar a atuao de agentes pblicos, milicianos e criminosos para atrapalhar as investigaes do homicdio da vereadora. O pedido foi feito pela procuradora-geral da Repblica, Raquel Dodge , e anunciado nesta quinta-feira pelo ministro da Segurana Pblica, Raul Jungmann, a quem a PF  subordinada. Tambm foi determinado que PF d proteo a duas pessoas que prestaram depoimentos sobre o caso.

Nesta sexta-feira, o Sindicato dos Delegados da Polcia Civil (Sindipol) e a Associao dos Delegados de Polcia do Estado (Adepol) divulgaram uma nota conjunta de repdio em que contestam as declaraes do ministro da Segurana Pblica . A nota acusa o ministro de ser um poltico profissional, que atua de forma vaidosa sem ter qualquer qualificao para atuar em um ministrio essencialmente tcnico.

Em nota, Jungmann rebateu as crticas dos policiais do Rio. Ele afirmou que "em nenhum momento identificou os agentes pblicos" suspeitos de atrapalhar a investigao. Disse ainda que se limitou a informar o pedido feito pela Procuradoria-Geral da Repblica. Destacou tambm que no houve federalizao da investigao do assassinato em si. Na quinta-feira, ele j havia dito que haver dois inquritos. A apurao do homicdio continuar com a Polcia Civil. J a Polcia Federal vai checar as denncias de fraudes e falhas na investigao.
'''

wordcloud = WordCloud(stopwords=STOPWORDS,
                          background_color='white', 
                      max_words=300,
                      width=2100, height=1200
                         ).generate(texto)
plt.figure(figsize=(40,20))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
